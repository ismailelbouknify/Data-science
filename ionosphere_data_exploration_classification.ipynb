## Detecting / Treating outliers

```python
def detect_outliers(db):
   
    q1 = db.quantile(0.25)
    q3= db.quantile(0.75)
    
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
  
    db_pur=db[(db>=lwr_bound) & (db<=upr_bound)]
    return db_pur
db_pur = detect_outliers(df.iloc[:,:-1])
db_pur
```

```python
db_pur.isnull().sum()
```

```python
for i  in db_pur:
    if (db_pur[i].isnull().any()):
       db_pur[i].fillna(df[i].mean(),inplace=True)
```

```python
db_pur.isnull().sum()
```

```python
y=df.iloc[:,-1]
y[y=='b']=0
y[y=='g']=1
y=y.astype('int')
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn import metrics 
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(db_pur, y, test_size=0.3, random_state=0)
```

```python
# Create Decision Tree classifer object
knn = KNeighborsClassifier()
# Train Decision Tree Classifer
knn.fit(X_train,y_train)
```

```python
predictions = knn.predict(X_test)
print("Accuracy",metrics.accuracy_score(y_test,predictions))
print("F1",metrics.f1_score(y_test,predictions))
```
